```python
def find_key(obj, key):
    if isinstance(obj, dict):
        for k, v in obj.items():
            if k == key:
                yield v
            if isinstance(v, (dict, list)):
                yield from find_key(v, key)
    elif isinstance(obj, list):
        for item in obj:
            yield from find_key(item, key)
```

```python
# Convert pandas DataFrame to Spark DataFrame before using withColumn
from pyspark.sql import DataFrame
import pandas as pd

def get_batsman_details(matchid):

    results = get_scorecard(matchid)
    if not results or len(results) == 0:
        print(f"No scorecard data for matchid: {matchid}")
        return
    df = pd.DataFrame(results[0])
    df_batsman_list = []

    for i in range(len(df['batsman'])):
        temp_df = pd.DataFrame(df['batsman'][i])
        temp_df['inningsId'] = df['inningsId'][i] if 'inningsId' in df.columns else i + 1
        temp_df['matchId'] = matchid
        df_batsman_list.append(temp_df)

    df_batsman = pd.concat(df_batsman_list, ignore_index=True)
    # ... your code to create df_batsman (assumed pandas DataFrame)
    if isinstance(df_batsman, pd.DataFrame):
        df_batsman = spark.createDataFrame(df_batsman)
    
    table_schema = spark.table("batsman_bronze").schema

    for field in table_schema:
        if field.name in df_batsman.columns:
            df_batsman = df_batsman.withColumn(
                field.name,
                df_batsman[field.name].cast(field.dataType)
            )
    # df_batsman.write.mode("append").saveAsTable("batsman_bronze")

    batsman_bronze_count = spark.sql(f"select count(*) from batsman_bronze where matchid="+str(matchid))

    if batsman_bronze_count.first()[0] == 0:
        # spark.createDataFrame(df_batsman).write.mode("append").saveAsTable("batsman_bronze")
        df_batsman.write.mode("append").saveAsTable("batsman_bronze")
    else:
        spark.sql(f"delete from batsman_bronze where matchid="+str(matchid))
        # spark.createDataFrame(df_batsman).write.mode("append").saveAsTable("batsman_bronze")
        df_batsman.write.mode("append").saveAsTable("batsman_bronze")
```

```python
# Convert pandas DataFrame to Spark DataFrame before using withColumn
from pyspark.sql import DataFrame
import pandas as pd

def get_bowler_details(matchid):

    results = get_scorecard(matchid)
    df = pd.DataFrame(results[0])
    df_bowler_list = []

    for i in range(len(df['bowler'])):
        temp_df = pd.DataFrame(df['bowler'][i])
        temp_df['inningsId'] = df['inningsId'][i] if 'inningsId' in df.columns else i + 1
        temp_df['matchId'] = matchid
        df_bowler_list.append(temp_df)

    df_bowler = pd.concat(df_bowler_list, ignore_index=True)
    # ... your code to create df_bowler (assumed pandas DataFrame)
    if isinstance(df_bowler, pd.DataFrame):
        df_bowler = spark.createDataFrame(df_bowler)
    
    #Check if bowlers_bronze is present
    # if spark._jsparkSession.catalog().tableExists("bowlers_bronze") == False:
    #     df_bowler.write.mode("overwrite").saveAsTable("bowlers_bronze")

    table_exists = (
    spark.sql(
        "SHOW TABLES LIKE 'bowlers_bronze'"
    )
    .count() > 0
    )

    if not table_exists:
        df_bowler.write.mode("overwrite").saveAsTable("bowlers_bronze")


    table_schema = spark.table("bowlers_bronze").schema

    for field in table_schema:
        if field.name in df_bowler.columns:
            df_bowler = df_bowler.withColumn(
                field.name,
                df_bowler[field.name].cast(field.dataType)
            )
    # df_bowler.write.mode("append").saveAsTable("bowler_bronze")

    bowler_bronze_count = spark.sql(f"select count(*) from bowlers_bronze where matchid="+str(matchid))

    if bowler_bronze_count.first()[0] == 0:
        # spark.createDataFrame(df_bowler).write.mode("append").saveAsTable("bowlers_bronze")
        df_bowler.write.mode("append").saveAsTable("bowlers_bronze")
    else:
        spark.sql(f"delete from bowlers_bronze where matchid="+str(matchid))
        # spark.createDataFrame(df_bowler).write.mode("append").saveAsTable("bowlers_bronze")
        df_bowler.write.mode("append").saveAsTable("bowlers_bronze")
```
