```python
#Show top 3 scorers of the week

spark.sql("""select
  name,
  total_runs
from (
  select
    name,
    sum(runs) as total_runs,
    row_number() over (
      order by sum(runs) desc
    ) as rn
  from batsman_bronze
  group by name
)
where rn < 4""").display()
```

```python
#Show top 3 wicket takers of the week

spark.sql("""select name,total_wickets from (
  select
    name,
    sum(wickets) as total_wickets,
    row_number() over (
      order by sum(wickets) desc
    ) as rn
  from bowlers_bronze
  group by name
)
where rn < 4  """).display()
```
