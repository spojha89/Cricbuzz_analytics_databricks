```python
import http.client

conn = http.client.HTTPSConnection("unofficial-cricbuzz.p.rapidapi.com")

headers = { 'x-rapidapi-host': "unofficial-cricbuzz.p.rapidapi.com" ,'x-rapidapi-key': "126bc65e90mshbeff1955b5e3b58p10d918jsn747ac022558a"}

conn.request("GET", "/matches/list?matchState=recent", headers=headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))
```

```python
#Find details of a series "seriesId": 9638 from json above

import json
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql.functions import from_json, col, unbase64, base64, split
from pyspark.sql.types import StructField, StructType, BooleanType, StringType, IntegerType, FloatType, LongType, ArrayType, DateType

spark = SparkSession.builder.appName("SparkStreaming").getOrCreate()


df=pd.json_normalize(json.loads(data.decode("utf-8")))
df



```

```python
def find_key(obj, key):
    if isinstance(obj, dict):
        for k, v in obj.items():
            if k == key:
                yield v
            if isinstance(v, (dict, list)):
                yield from find_key(v, key)
    elif isinstance(obj, list):
        for item in obj:
            yield from find_key(item, key)

```

```python
print(data.decode("utf-8"))
```

```python
json_data=json.loads(data.decode("utf-8"))
results = list(find_key(json_data, "matchInfo"))
print(results)
```

```python
print(type(results))
```

```python
#Convert list to pandas dataframe

df_matches=pd.DataFrame(results)
df_matches
#Convert pandas dataframe to spark dataframe
```

```python
#Convert date fields to date format

import pandas as pd

df_matches['startDate'] = pd.to_datetime(df_matches['startDate'], unit='ms')
df_matches['endDate'] = pd.to_datetime(df_matches['endDate'], unit='ms')
df_matches['seriesStartDt'] = pd.to_datetime(df_matches['seriesStartDt'], unit='ms')
df_matches['seriesEndDt'] = pd.to_datetime(df_matches['seriesEndDt'], unit='ms')
df_matches
```

```python
#convert pandas dataframe to spark dataframe

df_matches_spark=spark.createDataFrame(df_matches)
display(df_matches_spark)
```

```python
#Load to bronze delta table
# df_matches_spark.to_spark().write.mode("append").saveAsTable("bronze_matches")
df_matches_spark.write.mode("append").saveAsTable("bronze_matches")
```
